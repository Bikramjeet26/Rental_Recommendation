```python
uniq_names = list(df['name'].unique())
name_rus_to_eng = {}

for idx,name in enumerate(uniq_names):
    print(f'Done till {idx}')
    try:
        eng_name = translator.translate(name)
        name_rus_to_eng[name] = eng_name
    except Exception as e:
        print(f'No translation for name: {name}')
        name_rus_to_eng[name] = 'No_translation'
```

```python
all_cats = set(df['main_category'].unique())

for cats in df['categories']:
    for x in cats:
        all_cats.add(x)


uniq_main_cats = list(all_cats)
cat_rus_to_eng = {}

for idx,cat in enumerate(uniq_main_cats):
    print(f'Done till {idx}')
    try:
        eng_cat = translator.translate(cat)
        cat_rus_to_eng[cat] = eng_cat
    except Exception as e:
        print(f'No translation for category: {cat}')
        cat_rus_to_eng[cat] = 'No_translation'
```

```python
df['name_translated'] = df['name'].map(name_rus_to_eng)
df['main_category_translated'] = df['main_category'].map(cat_rus_to_eng)
```

```python
df['categories_translated'] = df['categories'].apply(lambda x: [cat_rus_to_eng[y] for y in x])
```

```python
def return_string_from_list(list_cats):
    string = ''
    if len(list_cats) != 0:
        string = 'Remaining_Categories: '
        for cat in list_cats:
            string += cat + ', '

    return string
```

```python
prod_desc = []
for idx, row in df.iterrows():
    string = 'Name: ' + row['name_translated'] + '\n' +  'Main_Category: ' + row['main_category_translated'] + '\n' + return_string_from_list(row['categories_translated'])
    prod_desc.append(string.strip())
```

```python
df['Description_Translated'] = prod_desc
df
```

```python
df[['id','Description_Translated']].to_csv('Product_ID_To_Descriptions.csv', index = False)
```
